```python
import os
from pathlib import Path
import torch
from utils import compute_mel_spectrogram

# Define paths
input_dir = Path("data/techniques")  # Folder with .wav files
output_dir = Path("data/spectrograms")  # Where .pt files will be saved
output_dir.mkdir(parents=True, exist_ok=True)

# Process each .wav file
for file in input_dir.glob("*.wav"):
    try:
        print(f"Processing {file.name}")
        mel_tensor = compute_mel_spectrogram(file)

        # Save as .pt file
        save_path = output_dir / (file.stem + ".pt")
        torch.save(mel_tensor, save_path)

    except Exception as e:
        print(f"Error processing {file.name}: {e}")
```

```python
import matplotlib.pyplot as plt

spec = torch.load("data/spectrograms/Arpeggios_eb_minor.pt")

plt.figure(figsize=(10, 4))
plt.imshow(spec, aspect='auto', origin='lower')
plt.title("Mel-Spectrogram")
plt.xlabel("Time Frames")
plt.ylabel("Mel Bins")
plt.colorbar(format="%+2.0f dB")
plt.show()
```
